#1. Análise do faturamento


```python
import pandas as pd

# URLs dos dados
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Lendo os dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Calculando o faturamento total de cada loja
faturamento_loja1 = loja1['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Exibindo os resultados com separador de milhar (formato brasileiro)
print("Faturamento da Loja 1: R$ {:,.2f}".format(faturamento_loja1).replace(',', 'v').replace('.', ',').replace('v', '.'))
print("Faturamento da Loja 2: R$ {:,.2f}".format(faturamento_loja2).replace(',', 'v').replace('.', ',').replace('v', '.'))
print("Faturamento da Loja 3: R$ {:,.2f}".format(faturamento_loja3).replace(',', 'v').replace('.', ',').replace('v', '.'))
print("Faturamento da Loja 4: R$ {:,.2f}".format(faturamento_loja4).replace(',', 'v').replace('.', ',').replace('v', '.'))
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Dados de faturamento
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [1534509.12, 1488459.06, 1464025.03, 1384497.58]

# Função para formatar os valores com separador de milhar (formato brasileiro)
def formatar_milhar(valor, _):
    return f'R$ {valor:,.2f}'.replace(',', 'v').replace('.', ',').replace('v', '.')

# Criando o gráfico
plt.figure(figsize=(8, 6))
barras = plt.bar(lojas, faturamentos, color='skyblue')

# Adicionando valores no topo das barras
for barra, valor in zip(barras, faturamentos):
    plt.text(barra.get_x() + barra.get_width() / 2,
             barra.get_height(),
             f'R$ {valor:,.2f}'.replace(',', 'v').replace('.', ',').replace('v', '.'),
             ha='center', va='bottom', fontsize=10)

# Título e eixos
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(formatar_milhar))
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

A loja 4 é a menor em faturamento total.

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das planilhas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Lendo os dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Remover espaços nos nomes das colunas
loja1.columns = loja1.columns.str.strip()
loja2.columns = loja2.columns.str.strip()
loja3.columns = loja3.columns.str.strip()
loja4.columns = loja4.columns.str.strip()

# Contando o número de produtos por categoria de forma correta
produtos_loja1 = loja1.groupby("Categoria do Produto")["Produto"].count().sort_values(ascending=False)
produtos_loja2 = loja2.groupby("Categoria do Produto")["Produto"].count().sort_values(ascending=False)
produtos_loja3 = loja3.groupby("Categoria do Produto")["Produto"].count().sort_values(ascending=False)
produtos_loja4 = loja4.groupby("Categoria do Produto")["Produto"].count().sort_values(ascending=False)

# Mostrar os resultados
print("Loja 1 - Quantidade de produtos por categoria:")
print(produtos_loja1)
print("\nLoja 2 - Quantidade de produtos por categoria:")
print(produtos_loja2)
print("\nLoja 3 - Quantidade de produtos por categoria:")
print(produtos_loja3)
print("\nLoja 4 - Quantidade de produtos por categoria:")
print(produtos_loja4)

# Criando gráfico de barras para cada loja
fig, axs = plt.subplots(2, 2, figsize=(14, 10))  # 2x2 gráfico para 4 lojas

# Loja 1
bars = axs[0, 0].barh(produtos_loja1.index, produtos_loja1.values, color='skyblue')
axs[0, 0].set_title('Loja 1 - Produtos por categoria')
axs[0, 0].set_xlabel('Quantidade de produtos')

# Adicionando os valores nas barras
for bar in bars:
    axs[0, 0].text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
                   f'{bar.get_width():,.0f}', va='center', ha='left')

# Loja 2
bars = axs[0, 1].barh(produtos_loja2.index, produtos_loja2.values, color='lightgreen')
axs[0, 1].set_title('Loja 2 - Produtos por categoria')
axs[0, 1].set_xlabel('Quantidade de produtos')

# Adicionando os valores nas barras
for bar in bars:
    axs[0, 1].text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
                   f'{bar.get_width():,.0f}', va='center', ha='left')

# Loja 3
bars = axs[1, 0].barh(produtos_loja3.index, produtos_loja3.values, color='lightcoral')
axs[1, 0].set_title('Loja 3 - Produtos por categoria')
axs[1, 0].set_xlabel('Quantidade de produtos')

# Adicionando os valores nas barras
for bar in bars:
    axs[1, 0].text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
                   f'{bar.get_width():,.0f}', va='center', ha='left')

# Loja 4
bars = axs[1, 1].barh(produtos_loja4.index, produtos_loja4.values, color='lightyellow')
axs[1, 1].set_title('Loja 4 - Produtos por categoria')
axs[1, 1].set_xlabel('Quantidade de produtos')

# Adicionando os valores nas barras
for bar in bars:
    axs[1, 1].text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
                   f'{bar.get_width():,.0f}', va='center', ha='left')

# Ajuste de layout para não sobrepor os gráficos
plt.tight_layout()
plt.show()

```

# 3. Média de Avaliação das Lojas

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das planilhas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Lendo os dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Remover espaços nos nomes das colunas
loja1.columns = loja1.columns.str.strip()
loja2.columns = loja2.columns.str.strip()
loja3.columns = loja3.columns.str.strip()
loja4.columns = loja4.columns.str.strip()

# Média das avaliações por loja
media_avaliacao_loja1 = loja1["Avaliação da compra"].mean()
media_avaliacao_loja2 = loja2["Avaliação da compra"].mean()
media_avaliacao_loja3 = loja3["Avaliação da compra"].mean()
media_avaliacao_loja4 = loja4["Avaliação da compra"].mean()

# Garantir que a avaliação não ultrapasse 5
media_avaliacao_loja1 = min(media_avaliacao_loja1, 5)
media_avaliacao_loja2 = min(media_avaliacao_loja2, 5)
media_avaliacao_loja3 = min(media_avaliacao_loja3, 5)
media_avaliacao_loja4 = min(media_avaliacao_loja4, 5)

# Valores das médias das avaliações
medias_avaliacoes = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

# Criando gráfico de barras
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
plt.figure(figsize=(8, 6))

# Gráfico de barras
bars = plt.bar(lojas, medias_avaliacoes, color=['skyblue', 'lightgreen', 'lightcoral', 'lightyellow'])

# Título e rótulos
plt.xlabel('Lojas')
plt.ylabel('Média das Avaliações')
plt.title('Média das Avaliações por Loja')

# Adicionando os valores nas barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'{yval:.2f}', ha='center', va='bottom')

# Exibindo o gráfico
plt.ylim(0, 5)  # Ajuste do limite para as avaliações (0 a 5)
plt.show()

```

# 5. Frete Médio por Loja


```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das planilhas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Lendo os dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Remover espaços nos nomes das colunas
loja1.columns = loja1.columns.str.strip()
loja2.columns = loja2.columns.str.strip()
loja3.columns = loja3.columns.str.strip()
loja4.columns = loja4.columns.str.strip()

# Calcular o frete médio por loja
frete_medio_loja1 = loja1["Frete"].mean()
frete_medio_loja2 = loja2["Frete"].mean()
frete_medio_loja3 = loja3["Frete"].mean()
frete_medio_loja4 = loja4["Frete"].mean()

# Exibindo os valores de frete médio
print(f"Frete médio da Loja 1: R$ {frete_medio_loja1:.2f}")
print(f"Frete médio da Loja 2: R$ {frete_medio_loja2:.2f}")
print(f"Frete médio da Loja 3: R$ {frete_medio_loja3:.2f}")
print(f"Frete médio da Loja 4: R$ {frete_medio_loja4:.2f}")

# Criando gráfico de barras para comparar os fretes médios
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes_medios = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

plt.figure(figsize=(8, 6))
bars = plt.bar(lojas, fretes_medios, color=['skyblue', 'lightgreen', 'lightcoral', 'lightyellow'])
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.title('Frete Médio por Loja')

# Adicionando os valores nas barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'R$ {yval:.2f}', ha='center', va='bottom')

plt.show()
```
